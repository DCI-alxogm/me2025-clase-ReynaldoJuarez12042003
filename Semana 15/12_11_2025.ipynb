```python
from scipy.optimize import linprog
```

```python
c=[-150,-175]
A_ub=[[7,11],[10,8], [1,0],[0,1]]
b_ub=[77,80,9,6]
x1_bounds=(0,None)
x2_bounds=(0,None)
bounds=[x1_bounds,x2_bounds]
result=linprog(c,A_ub=A_ub, b_ub=b_ub, bounds=bounds, method='highs')
```
